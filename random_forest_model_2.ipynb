```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(10)
```

# Select your features (columns)

```python
X = df.drop(["koi_disposition", "koi_steff", "koi_srad", "koi_slogg", "koi_slogg_err1", "koi_srad_err2", "koi_tce_plnt_num"], axis=1)
y = df["koi_disposition"]
y_names = ["CONFIRMED", "FALSE POSITITVE", "CANDIDATE"]
feature_names = X.columns
print(X.shape, y.shape)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
from sklearn import tree
clf_model = tree.DecisionTreeClassifier()
clf_model = clf_model.fit(X_train_scaled, y_train)
clf_model.score(X_test_scaled, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=900)
rf_model = rf_model.fit(X_train_scaled, y_train)
rf_model.score(X_test_scaled, y_test)
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth': [3, 5, 10],
            'min_samples_split': [2, 5, 10],
             'n_estimators': [850, 900, 950],
             'max_features' : [30, 35, 40] }

grid = GridSearchCV(rf_model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train.ravel())
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
